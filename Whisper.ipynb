```python
import openai
```

```python
import os
openai.api_key = os.getenv('OPEN_AI_KEY') # captura a API-key salva no ambiente de sistema
```

```python
audio_file = open("Y:\\4k downloader\\Flagamer S03E16 - F1 Championship Edition vs F1 2013.mp3", "rb")
```

```python
transcript = openai.Audio.transcribe("whisper-1", audio_file, response_format="verbose_json")
print(transcript)
```

```python
transcript['text']
```

```python
# Abre o arquivo em modo de escrita ('w' significa write)
with open('roteiro.txt', 'w') as arquivo:
    # Escreve o texto no arquivo
    arquivo.write(f'{transcript["text"]}')

# O arquivo é automaticamente fechado quando o bloco 'with' é encerrado
```

```python
for segment in transcript['segments']:
    start_minutes, start_seconds = divmod(segment['start'], 60)
    timestamp = f"{round(start_minutes):02d}:{round(start_seconds):02d}"
    text = segment["text"]
    
    print(f"{timestamp} - {text}")
```
